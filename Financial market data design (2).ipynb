```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
```

```python
start = datetime.datetime(2006, 1, 1)
end = datetime.datetime(2016, 1, 1)

# Bank of America
BAC = data.DataReader("BAC", 'yahoo', start, end)

# CitiGroup
C = data.DataReader("C", 'yahoo', start, end)

# Goldman Sachs
GS = data.DataReader("GS", 'yahoo', start, end)

# JPMorgan Chase
JPM = data.DataReader("JPM", 'yahoo', start, end)

# Morgan Stanley
MS = data.DataReader("MS", 'yahoo', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'yahoo', start, end)
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis=1, keys=tickers)
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

```python
returns = pd.DataFrame()
for tick in tickers:
    returns[tick + 'Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head(50)
```

```python
import seaborn as sns
```

** distplot usando seaborn dos retornos de 2015 para Morgan Stanley **

```python
sns.distplot(returns[(returns.index.date >= datetime.date(2015, 1, 1)) & (returns.index.date < datetime.date(2016, 1, 1))]['MSReturn'])
```

** Crie um distplot usando seaborn dos retornos de 2008 para CitiGroup **

```python
sns.distplot(returns[(returns.index.date >= datetime.date(2008, 1, 1)) & (returns.index.date < datetime.date(2009, 1, 1))]['CReturn'])
```

*grfico de linha mostrando o preo de fechamento para cada banco para todo o ndice de tempo *

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
for tick in tickers:
    bank_stocks[tick]['Close'].plot(figsize=(12, 4), label=tick)
plt.legend()
```

```python
bank_stocks.xs(key='Close', level='Stock Info', axis=1).plot()
```

*mdia de 30 dias para o preo prximo do Bank Of America para o ano de 2008 *

```python
plt.figure(figsize=(12, 6))
BAC2008 = BAC['Close'][BAC.index.year == 2008]
BAC2008_media_movel = BAC2008.rolling(window=30).mean().plot(label='Mdia mvel de 30 dias')
BAC2008.plot(label='Fechamento')
plt.legend()
```

*mapa de calor da correlao entre os preos de fechamento das aes*

```python
corr_matrix = bank_stocks.xs(key='Close',level='Stock Info', axis=1).corr()
```

```python
sns.heatmap(corr_matrix, annot=True)
```
